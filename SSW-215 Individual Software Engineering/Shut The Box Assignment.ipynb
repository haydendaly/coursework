```python
import random    #for rolling the die
import sys       #for ending code

win = ["Down","Down","Down","Down","Down","Down","Down","Down","Down"]
player1 = ["Up", "Up", "Up", "Up", "Up", "Up", "Up", "Up", "Up"]
player2 = ["Up", "Up", "Up", "Up", "Up", "Up", "Up", "Up", "Up"]

def current_hand(player, i):    #displays current hand
    x = 1
    print("Your current hand for player",i,"is:\n")
    while x <= 9:
        print("    ",x,":",player[x-1])
        x += 1
    print("\n")
    
def flip(player, dieNum):    #turns tiles over
    player[dieNum-1] = "Down"

def roll(player, i):    #simulates the rolling of dice and decision of player
    current_hand(player, i)
    die1 = random.randint(1,6)    #random int on range of 1 to 6
    die2 = random.randint(1,6)
    if player[8] == "Down" and player[7] == "Down" 
                    and player[6] == "Down" and player[5] == "Down":
        print("Your die rolled",die1,"/n")
    else:
        print("Your die rolled",die1,"and",die2,"\n")
        if (die1 + die2) >= 10:    #would create tile not in set so omits choice
            flip(player, die1)
            flip(player, die2)
        else:
            print("Would you like to flip",die1,"and",die2,
                  "or just",die1+die2,"?\n(Ex: Type",die1,
                  "for",die1,"and",die2,"or",die1+die2,
                  "for",die1+die2,")")
            user_input = input("")
            if int(user_input) == die1:    #player chooses both dice
                flip(player, die1)
                flip(player, die2)
            else:                          #player chooses sum of dice
                flip(player, die1+die2)
    current_hand(player, i)
            
                
while player1 != win and player2 != win:
    print("PLAYER 1:")
    roll(player1, 1)    #1 (or i in function) makes it easier to reference the player
    if player1 == win:
        break
    else:
        print("PLAYER 2:")
        roll(player2, 2)
    
if player1 == win:
    print("Player 1 Wins!")
else:
    print("Player 2 Wins!")
sys.exit(0)
```
